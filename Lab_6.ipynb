```python
import boto3
import pandas as pd
```

```python
comprehend = boto3.client('comprehend', region_name='ap-south-1')
```

```python
texts = [
    "I really enjoyed using this new phone, the camera is fantastic!",
    "The delivery was late and the packaging was damaged.",
    "The movie was okay, not great but not terrible either.",
    "I'm so happy with the customer service, they solved my issue quickly!"
]
```

```python
result = comprehend.detect_sentiment(Text=texts[0], LanguageCode='en')
print(result)
```

```python
batch_result = comprehend.batch_detect_sentiment(TextList=texts, LanguageCode='en')
for i, res in enumerate(batch_result['ResultList']):
    print(f"Text: {texts[i]}")
    print(f"Detected Sentiment: {res['Sentiment']}")
    print(f"Confidence Scores: {res['SentimentScore']}")
    print("-" * 60)
```
